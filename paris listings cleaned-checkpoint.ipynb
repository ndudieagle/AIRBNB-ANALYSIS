```python
#Filter the data down to rows where the city is Paris, and keep only the columns ‘host_since’, ‘neighbourhood’, ‘city’, ‘accommodates’, and ‘price’


import pandas as pd

# Load the data
listings = pd.read_csv("listings.csv", encoding="ISO-8859-1", low_memory=False)

# Filter rows where the city is Paris
paris_listings = listings[listings['city'].str.lower() == 'paris']

# Select only the required columns
paris_listings = paris_listings[['host_since', 'neighbourhood', 'city', 'accommodates', 'price']]

# Save the filtered data to a new CSV file (optional)
paris_listings.to_csv("paris_listings_filtered.csv", index=False)

# Display the filtered data
print(paris_listings.head())
```

```python
#QA the Paris listings data: check for missing values, and calculate the minimum, maximum, and average for each numeric field

import pandas as pd
import numpy as np  # Import numpy

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Check for missing values
missing_values = paris_listings.isnull().sum()
print("Missing Values:\n", missing_values)

# Step 2: Calculate summary statistics for numeric fields
# Select numeric columns
numeric_columns = paris_listings.select_dtypes(include=[np.number]).columns

# Calculate min, max, and average for numeric fields
summary_stats = paris_listings[numeric_columns].agg(['min', 'max', 'mean'])
print("\nSummary Statistics for Numeric Fields:\n", summary_stats)
```

```python

#Create a table named paris_listings_neighbourhood that groups Paris listings by 'neighbourhood' and calculates the mean price (sorted low to high)

import pandas as pd

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Group by 'neighbourhood' and calculate the mean price
paris_listings_neighbourhood = (
    paris_listings.groupby('neighbourhood')['price']
    .mean()
    .reset_index()
    .sort_values(by='price', ascending=True)  # Sort by mean price (low to high)
)

# Rename the 'price' column to 'mean_price' for clarity
paris_listings_neighbourhood = paris_listings_neighbourhood.rename(columns={'price': 'mean_price'})

# Save the result to a CSV file (optional)
paris_listings_neighbourhood.to_csv("paris_listings_neighbourhood.csv", index=False)

# Display the result
print(paris_listings_neighbourhood)
```

```python
#Create a table named paris_listings_accomodations, filter down to the most expensive neighborhood, 
#group by the ‘accommodations’ column, and add the mean price for each value of ‘accommodates’ (sorted low to high)

import pandas as pd

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Identify the most expensive neighborhood
# Calculate the mean price for each neighborhood
neighbourhood_mean_prices = (
    paris_listings.groupby('neighbourhood')['price']
    .mean()
    .reset_index()
)

# Find the neighborhood with the highest mean price
most_expensive_neighbourhood = neighbourhood_mean_prices.loc[
    neighbourhood_mean_prices['price'].idxmax(), 'neighbourhood'
]

# Step 2: Filter the data to the most expensive neighborhood
most_expensive_listings = paris_listings[
    paris_listings['neighbourhood'] == most_expensive_neighbourhood
]

# Step 3: Group by 'accommodates' and calculate the mean price
paris_listings_accommodations = (
    most_expensive_listings.groupby('accommodates')['price']
    .mean()
    .reset_index()
    .sort_values(by='accommodates', ascending=True)  # Sort by 'accommodates' (low to high)
)

# Rename the 'price' column to 'mean_price' for clarity
paris_listings_accommodations = paris_listings_accommodations.rename(columns={'price': 'mean_price'})

# Save the result to a CSV file (optional)
paris_listings_accommodations.to_csv("paris_listings_accommodations.csv", index=False)

# Display the result
print(f"Most Expensive Neighborhood: {most_expensive_neighbourhood}")
print(paris_listings_accommodations)
```

```python
#Create a table called paris_listings_over_time grouped by the ‘host_since’ year, and calculate the average price and
#count of rows representing the number of new hosts

import pandas as pd

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Extract the year from the 'host_since' column
paris_listings['host_since'] = pd.to_datetime(paris_listings['host_since'], errors='coerce')
paris_listings['host_since_year'] = paris_listings['host_since'].dt.year

# Step 2: Group by the 'host_since_year' and calculate the average price and count of rows
paris_listings_over_time = (
    paris_listings.groupby('host_since_year')
    .agg(
        average_price=('price', 'mean'),  # Calculate the average price
        new_hosts=('host_since_year', 'count')  # Count the number of new hosts
    )
    .reset_index()
    .sort_values(by='host_since_year', ascending=True)  # Sort by year (ascending)
)

# Save the result to a CSV file (optional)
paris_listings_over_time.to_csv("paris_listings_over_time.csv", index=False)

# Display the result
print(paris_listings_over_time)
```

```python
#Create a horizontal bar chart of the average price by neighborhood in Paris, and make sure to add a title and change axis labels as needed

import pandas as pd
import matplotlib.pyplot as plt

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Group by 'neighbourhood' and calculate the mean price
paris_listings_neighbourhood = (
    paris_listings.groupby('neighbourhood')['price']
    .mean()
    .reset_index()
    .sort_values(by='price', ascending=True)  # Sort by mean price (low to high)
)

# Step 2: Create a horizontal bar chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.barh(paris_listings_neighbourhood['neighbourhood'], paris_listings_neighbourhood['price'], color='skyblue')

# Step 3: Add a title and axis labels
plt.title('Average Price by Neighborhood in Paris', fontsize=16)
plt.xlabel('Average Price (€)', fontsize=12)
plt.ylabel('Neighborhood', fontsize=12)

# Step 4: Add grid lines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Step 5: Display the chart
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
#Create two line charts: one showing the count of new hosts over time, 
#and one showing average price. Set the y-axis limit to 0, add a title, and change axis labels as needed
import pandas as pd
import matplotlib.pyplot as plt

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Extract the year from the 'host_since' column
paris_listings['host_since'] = pd.to_datetime(paris_listings['host_since'], errors='coerce')
paris_listings['host_since_year'] = paris_listings['host_since'].dt.year

# Step 2: Group by the 'host_since_year' and calculate the count of new hosts and average price
paris_listings_over_time = (
    paris_listings.groupby('host_since_year')
    .agg(
        new_hosts=('host_since_year', 'count'),  # Count of new hosts
        average_price=('price', 'mean')  # Average price
    )
    .reset_index()
    .sort_values(by='host_since_year', ascending=True)  # Sort by year (ascending)
)

# Step 3: Create the first line chart (Count of New Hosts Over Time)
plt.figure(figsize=(12, 6))  # Set the figure size

# Plot the count of new hosts
plt.subplot(1, 2, 1)  # Create the first subplot
plt.plot(paris_listings_over_time['host_since_year'], paris_listings_over_time['new_hosts'], marker='o', color='blue', linestyle='-')
plt.title('Count of New Hosts Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of New Hosts', fontsize=12)
plt.ylim(0, paris_listings_over_time['new_hosts'].max() + 10)  # Set y-axis limit to start at 0
plt.grid(linestyle='--', alpha=0.7)

# Step 4: Create the second line chart (Average Price Over Time)
plt.subplot(1, 2, 2)  # Create the second subplot
plt.plot(paris_listings_over_time['host_since_year'], paris_listings_over_time['average_price'], marker='o', color='green', linestyle='-')
plt.title('Average Price Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Price (€)', fontsize=12)
plt.ylim(0, paris_listings_over_time['average_price'].max() + 50)  # Set y-axis limit to start at 0
plt.grid(linestyle='--', alpha=0.7)

# Step 5: Adjust layout and display the charts
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
#Create a dual axis line chart to show both new hosts and average price over time
import pandas as pd
import matplotlib.pyplot as plt

# Load the filtered Paris listings data
paris_listings = pd.read_csv("paris_listings_filtered.csv")

# Step 1: Extract the year from the 'host_since' column
paris_listings['host_since'] = pd.to_datetime(paris_listings['host_since'], errors='coerce')
paris_listings['host_since_year'] = paris_listings['host_since'].dt.year

# Step 2: Group by the 'host_since_year' and calculate the count of new hosts and average price
paris_listings_over_time = (
    paris_listings.groupby('host_since_year')
    .agg(
        new_hosts=('host_since_year', 'count'),  # Count of new hosts
        average_price=('price', 'mean')  # Average price
    )
    .reset_index()
    .sort_values(by='host_since_year', ascending=True)  # Sort by year (ascending)
)

# Step 3: Create a dual-axis line chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the count of new hosts on the primary y-axis
color = 'blue'
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Number of New Hosts', fontsize=12, color=color)
ax1.plot(paris_listings_over_time['host_since_year'], paris_listings_over_time['new_hosts'], marker='o', color=color, linestyle='-', label='New Hosts')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, paris_listings_over_time['new_hosts'].max() + 10)  # Set y-axis limit for new hosts

# Create a secondary y-axis for average price
ax2 = ax1.twinx()
color = 'green'
ax2.set_ylabel('Average Price (€)', fontsize=12, color=color)
ax2.plot(paris_listings_over_time['host_since_year'], paris_listings_over_time['average_price'], marker='o', color=color, linestyle='-', label='Average Price')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, paris_listings_over_time['average_price'].max() + 50)  # Set y-axis limit for average price

# Add a title
plt.title('New Hosts and Average Price Over Time in Paris', fontsize=16)

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Add grid lines
ax1.grid(linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```
